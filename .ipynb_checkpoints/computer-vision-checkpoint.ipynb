# Computer Vision Using Fastai

#### In this notebook we'll build a Learner and fine tune pre-trained model on computer vision tasks.

```python
from fastai.vision.all import *
```

#### Single label Classification

We'll use the Oxford-IIIT Pet Dataset to classifiy cats-vs-dogs. The dataset contains images of cats and dogs of 37 different breeds.

```python
path = untar_data(URLs.PETS)
```

```python
path.ls()
```

```python
# get image files recursively
files = get_image_files(path/"images")
```

```python
files[0]
```

```python
len(files)
```

```python
# label function
def label_func(f):
    return f[0].isupper()
```

```python
dls = ImageDataLoaders.from_name_func(path, files, label_func, item_tfms=Resize(224), num_workers=0)
```

```python
# True for cat and False for dogs
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.fine_tune(1)
```
